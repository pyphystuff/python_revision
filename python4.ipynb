# (Very) Simple Harmonic oscillator visualization
    

```python
#importing modules
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
#other useful modules
import numpy as np
import sympy as s
```

# Harmonic oscillator solution:
    
# $ x(t) = cos(wt+\alpha)$ 
# $w= \sqrt{(k/m)}$

    
    
    

```python
#lets define some constants
k=5
m=4
w= np.sqrt(k/m)
alpha= 15
```

```python
#Basic labeling
fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(-20, 20, 100)
y1 = np.cos(w*x +alpha)

y2 = np.cos(w*x +30)


ax.plot(x, y1)
ax.plot(x, y2)

ax.set_ylabel("y label")
ax.set_xlabel("x label")

```

```python
#lets damp this
fig, ax = plt.subplots(figsize=(8, 4))
#damping factor
d=0.09

x = np.linspace(-20, 20, 100)
y1 = np.exp(-d*x)*np.cos(w*x +alpha)

y2 = np.exp(-d*x)*np.cos(w*x +30)
y4 = np.exp(-d*x)
y3 = -np.exp(-d*x)

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)
ax.plot(x, y4)

ax.set_ylabel("y label")
ax.set_xlabel("x label")
```

Lets work with some energy 

Potencial energy

u=(1/2)kx^2

Kinetic energy

u=(1/2)mv^2


```python

fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(-10, 10, 100)
#potential energy

y1 = (1/2)*k*(np.cos(w*x +alpha))**2

#kinetic energy

y2 = (1/2)*m*w*(np.sin(w*x +alpha))**2


ax.plot(x, y1)
ax.plot(x, y2)


ax.set_ylabel("y label")
ax.set_xlabel("x label")
```

```python
#space phase



fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(-10, 10, 100)
#potential energy

y1 = np.cos(w*x +alpha)

#kinetic energy

y2 = w*np.sin(w*x +alpha) #velocity


#ax.plot(x, y1)
ax.plot(y1, y2)


ax.set_ylabel("y label")
ax.set_xlabel("x label")
```
