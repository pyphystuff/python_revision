```python
#importing modules
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
```

```python
#other useful modules
import numpy as np
import sympy as s
```

```python
#Basic labeling
fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(-20, 20, 100)
y = np.sin(x) / x

ax.plot(x, y)

ax.set_ylabel("y label")
ax.set_xlabel("x label")

```

```python
#lets combine some exercices: a figure with labels and backgorund color


#changing some colors with facecolor: 
fig = plt.figure(figsize=(8, 2.5), facecolor="#eafff5")
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes((left, bottom, width, height),
         facecolor="#17becf") #changed color here also

x = np.linspace(-10,10,100)
y = np.sin(x)*np.exp(-x)


ax.plot(x, y, color='red', label ="y=sin(x)*exp(x)")


ax.set_ylabel("y label")
ax.set_xlabel("x label")

ax.legend()   #always call the legend

```

```python
#latex format is available

fig, ax = plt.subplots(figsize=(12, 3))

ax.set_yticks([])
ax.set_xticks([])
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-0.05, 0.25)
ax.axhline(0)

ax.text(0, 0.1, "Text label", fontsize=14, family="serif")

ax.plot(1, 0, 'o')
ax.annotate("Annotation",
            fontsize=14, family="serif",
            xy=(1, 0), xycoords='data',
            xytext=(+20, +50), textcoords='offset points', 
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.5"))

ax.text(2, 0.1, r"Equation: $i\hbar\partial_t \Psi = \hat{H}\Psi$", fontsize=14, family="serif")

fig.savefig("ch4-text-annotation.pdf")
```
